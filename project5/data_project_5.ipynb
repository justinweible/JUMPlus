#### 17. How many people have a credit card that expires in 2020?

```python
purchases_df.dtypes
```

```python
purchases_df['CC Exp Date'] = pd.to_datetime(purchases_df['CC Exp Date'], format='%m/%y')
```

```python
purchases_df['CC Exp Date'] = purchases_df['CC Exp Date'].dt.strftime('%m/%Y')
purchases_df.head(10)
```

```python
purchases_df['Year'] = purchases_df['CC Exp Date'].str.split('/').str[-1]
purchases_df.head(10)
```

```python
purchases_df['Year'].value_counts()
```

```python
year_df = purchases_df.loc[purchases_df['Year']=='2020']
```

```python
year_count = year_df['Year'].count()
print("Number of people who have credit cards that expire in 2020: {}".format(year_count))
```
